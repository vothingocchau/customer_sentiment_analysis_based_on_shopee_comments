> * Cn vi nhm khch hng nhm negative, n b lch phi nhng khng qu nhiu. Nu y l cc feature m ta quan tm trong lc model training, ta c th p dng log-normalization ln chng trc sau  training sau.

Tip theo, ta s kim chng mi quan h gia bin phn loi `label` v bin lin tc `length` v `no_words` vi **null hypothesis** l gia cc class trong `label` chu tc ng t `length` hoc `no_words` l nh sau. Ta s p dng ANOVA one-way  kim tra iu ny.

```python
Detective.onewayANOVA(reviews, 'length')
```

> **Nhn xt**:
> * V gi tr **p-value** (`PR(>F)`) nh hn 5% tc ta phi bc b **null hypothesis**.
> * Tuy nhin, ta ch bit l `label` chu tc ng khng u gia cc class bi `length`, nhng ta khng bit c class no s chu tc ng ln hn hoc km hn so vi cc class cn li, buc tip theo ta s p dng **post-hoc testing**  kim chng iu ny.
> * Ta s p dng **t-test**  kim chng iu ny, tuy nhin do s class trong `label` ch c hai, lm iu ny l hi tha thi, v cc s  boxplot  pha trn cng  cho ta thy c gia cc class ca `label` l c s khc bit ng k v gi tr trung bnh, tc c class chu tc ng ln hn bi `length`, ngi ta thng p dng **post-hoc testing** khi c t 3 class trong categorical variable tr ln, nhng ta c p dng th xem mi th pha trc ta lm cho n by gi bao gm trc quan ha, thng k m t c ng trn hypothesis testing khng.

```python
Detective.onewayANOVA(reviews, 'length', True)
```

> **Nhn xt**:
> * R rng,  trng `reject-hs` vi gi tr **True** tc l c s khc bit v gi tr trung bnh ng k bi `length` ln cc class trong `label`.
> * Tip theo, ta cng c th p dng tng t cho cp feature `label` v `no_words` mc d ta  bit trc l cng cho ra kt lun tng t nh cp `label` v `length` thi. V  bc trc quan ta bit c rng cng nhiu t th bnh lun cng di.

```python
Detective.onewayANOVA(reviews, 'no_words')
```

```python
Detective.onewayANOVA(reviews, 'no_words', True)
```
